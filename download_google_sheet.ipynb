```python
from pathlib import Path
from playwright.sync_api import sync_playwright

SHEET_URL = "https://docs.google.com/spreadsheets/d/1QYvhHhgrBxwuumcbLZvtnKTvVi61RA5TeKZ6p_cAXSY/edit?gid=990977647#gid=990977647"

download_dir = Path("downloads")
download_dir.mkdir(exist_ok=True)

with sync_playwright() as p:
    browser = p.chromium.launch_persistent_context(
        user_data_dir="google_login_profile",
        headless=False,
        accept_downloads=True
    )

    page = browser.new_page()
    page.goto(SHEET_URL, wait_until="networkidle")

    page.keyboard.press("Alt+f")
    page.keyboard.press("d")
    page.keyboard.press("x")

    download = page.wait_for_download(timeout=60000)
    path = download_dir / "in_transit_raw.xlsx"
    download.save_as(path)

    browser.close()

print(f"Downloaded: {path}")
```
